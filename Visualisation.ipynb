```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\USER\Documents\Restaurants.csv")
df.head(3)
```

```python
cuisine_city = pd.DataFrame(df.groupby("City")["Aggregate rating"].mean().reset_index())
cuisine_city = cuisine_city.sort_values(by = "Aggregate rating",ascending = False).head(10)
```

```python
#Horizontal bar chart
plt.barh(cuisine_city["City"], cuisine_city["Aggregate rating"])
plt.xlabel('Aggregate Rating')
plt.ylabel('City')
plt.title('Average Ratings of Cities')
plt.show()
```

```python
textrating = pd.DataFrame(df.groupby("Rating text")["Votes"].mean().reset_index())
textrating = textrating[textrating["Rating text"] != "Not rated"]
```

```python
plt.pie(textrating["Votes"],labels = textrating["Rating text"],autopct='%1.1f%%')
plt.show()
```

```python
pricerangevotes = pd.DataFrame(df.groupby("Price range")["Votes"].mean().reset_index())
pricerangevotes
```

```python
plt.scatter(pricerangevotes["Price range"], pricerangevotes["Votes"],c = "green")
plt.xlabel("Price range")  
plt.ylabel("Votes")        
plt.title("Price Range vs Votes") 
plt.show()
```

```python
pricerangeaggrating = pd.DataFrame(df.groupby("Price range")["Aggregate rating"].mean().reset_index())
pricerangeaggrating
```

```python
plt.plot(pricerangeaggrating["Price range"],pricerangeaggrating["Aggregate rating"])
plt.xlabel("Price range")  
plt.ylabel("Aggregate rating")        
plt.title("Price Range vs Aggregate Rating") 
plt.show()
```
